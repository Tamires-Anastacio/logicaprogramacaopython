8- Faça um algoritmo que receba uma menção, podendo ser I, R, B ou MB. Se o
aluno tirou I ou R, está Reprovado, senão se o aluno tirou B ou MB está Aprovado

```python
# Seu código aqui

#entrada
nota=float(input("Digite a sua nota"))

#processamento
if nota<=5:
  print("Sua menção final é I")
elif nota>=5 and nota < 7:
  print("Sua menção final é R")
elif nota>=7 and nota < 8.8:
  print("Sua menção final é B")
elif nota>=8.8:
  print("Sua menção final é MB")

```
